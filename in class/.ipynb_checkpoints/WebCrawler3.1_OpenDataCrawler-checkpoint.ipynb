```python
#網路連線
import urllib .request as request
# src="https://www.cjcu.edu.tw/"   #將首頁資料存入data(16進制) note.標籤、屬性
# src = 'https://data.taipei/api/v1/dataset/dfe02544-72c9-4d2e-bd19-81276d0e7f9f?scope=resourceAquire'

src = 'https://duckjiang.wordpress.com/'

with request.urlopen(src)as response:
    data=response.read() #取得長榮大學網站的原始碼(HTML、CSS、JASON) 
print(data)
```

```python
with request.urlopen(src)as response:
    data=response.read().decode("utf-8") #utf-8解碼
print(data)
```

```python
#串接、擷以公開資料 
import urllib .request as request   #opendata>網頁API位址，results是陣列
src="https://data.taipei/api/v1/dataset/dfe02544-72c9-4d2e-bd19-81276d0e7f9f?scope=resourceAquire"
with request.urlopen(src)as response:
    data=response.read().decode("utf-8") 
print(data)
```

```python
import urllib .request as request
import json#使用JSON格式讀取
src="https://data.taipei/api/v1/dataset/dfe02544-72c9-4d2e-bd19-81276d0e7f9f?scope=resourceAquire"
with request.urlopen(src) as response:
    data=json.load(response)#使用JSON格式讀取   #跟上一段不同在於此段用json讀取
print(data)
```

```python
#將公司名稱列表出來 
clist = data["result"]["results"] 
with open("data.txt","w",encoding="utf-8") as file:
    for location in clist:
        file.write(location["稽查日期"],location["稽查地點"],location["行政區"],"\n")
        print(location["稽查日期"],location["稽查地點"],location["行政區"])
```
